```python
def map_genre(genre):
    if genre in ["Sequential Art"]:
        return "Graphic Novels & Comics"

    elif genre in [
        "Fiction","Contemporary","Novels","Adult Fiction","Womens Fiction",
        "Classics","Poetry","Historical Fiction"
    ]:
        return "Fiction"

    elif genre in ["Fantasy","Science Fiction","Paranormal","Horror"]:
        return "Speculative Fiction"

    elif genre in ["Mystery","Thriller","Crime","Suspense","Romance"]:
        return "Mystery & Romance"

    elif genre in ["Young Adult","New Adult","Children"]:
        return "Young Readers"

    elif genre in ["Nonfiction","Business","Politics","History",
                   "Biography","Autobiography","Historical"]:
        return "Non-Fiction"

    elif genre in [
        "Philosophy","Psychology","Religion","Spirituality",
        "Self Help","Christian","Christian Fiction"
    ]:
        return "Philosophy / Religion"

    elif genre in [
        "Art","Music","Food and Drink","Travel","Humor",
        "Sports and Games","Parenting","Cultural"
    ]:
        return "Lifestyle & Culture"

    elif genre in ["Academic","Science","Health"]:
        return "Academic"

    else:
        return "Other"
```

```python
df["genre_group"] = df["genre"].apply(map_genre)
df["genre_group"].value_counts()
```

```python
def map_genres(genre):
    # ---------- FICTION ----------
    if genre in [
        "Fiction","Contemporary","Novels","Adult Fiction","Womens Fiction"
    ]:
        return "Fiction", "Fiction (Core)"

    elif genre in ["Fantasy","Science Fiction","Paranormal","Horror"]:
        return "Fiction", "Speculative Fiction"

    elif genre in ["Mystery","Thriller","Crime","Suspense","Romance"]:
        return "Fiction", "Mystery & Romance"

    elif genre in ["Classics","Poetry","Historical Fiction"]:
        return "Fiction", "Classics & Literature"

    elif genre in ["Sequential Art"]:
        return "Fiction", "Graphic Novels & Comics"

    # ---------- NON-FICTION ----------
    elif genre in ["Nonfiction","Business","Politics"]:
        return "Non-Fiction", "Non-Fiction (General)"

    elif genre in ["History","Historical","Biography","Autobiography"]:
        return "Non-Fiction", "History & Biography"

    elif genre in [
        "Philosophy","Psychology","Religion","Spirituality",
        "Self Help","Christian","Christian Fiction"
    ]:
        return "Non-Fiction", "Philosophy / Religion"

    # ---------- LIFESTYLE ----------
    elif genre in [
        "Art","Music","Food and Drink","Travel","Humor",
        "Sports and Games","Parenting","Cultural"
    ]:
        return "Lifestyle & Culture", "Lifestyle & Culture"

    # ---------- YOUNG READERS ----------
    elif genre in ["Young Adult","New Adult","Children"]:
        return "Young Readers", "Young Readers"

    # ---------- ACADEMIC ----------
    elif genre in ["Academic","Science","Health"]:
        return "Academic", "Academic"

    # ---------- OTHER ----------
    else:
        return "Other", "Other"


df[["genre_macro", "genre_sub"]] = df["genre"].apply(
    lambda x: pd.Series(map_genres(x))
)
```

```python
def map_genre_custom(genre):
    # 1. Non-Fiction (General)
    if genre in [
        "Nonfiction", "Business", "Biography"
    ]:
        return "Non-Fiction (General)"

    # 2. History / Psychology / Human Stories (includes Sequential Art)
    elif genre in [
        "History", "Psychology", "Autobiography", "Self Help", "Sequential Art"
    ]:
        return "History / Psychology / Human Stories"

    # 3. Core Fiction
    elif genre in [
        "Fiction", "Mystery", "Historical Fiction",
        "Womens Fiction", "Science Fiction", "Christian Fiction"
    ]:
        return "Core Fiction"

    # 4. Speculative / Intense Fiction
    elif genre in [
        "Fantasy", "Horror", "Science", "Thriller"
    ]:
        return "Speculative / Intense Fiction"

    # 5. Lifestyle & Light Reads
    elif genre in [
        "Food and Drink", "Travel", "Humor", "Sports and Games", "Health"
    ]:
        return "Lifestyle & Light Reads"

    # 6. Young Adult / New Adult
    elif genre in [
        "Young Adult", "New Adult"
    ]:
        return "Young Adult / New Adult"

    # 7. Romance
    elif genre == "Romance":
        return "Romance"

    # 8. Children
    elif genre == "Childrens":
        return "Children"

    # 9. Arts / Philosophy / Spiritual
    elif genre in [
        "Classics", "Poetry", "Music",
        "Philosophy", "Art", "Religion", "Spirituality"
    ]:
        return "Arts / Philosophy / Spiritual"

    # 10. Other
    else:
        return "Other"
```
